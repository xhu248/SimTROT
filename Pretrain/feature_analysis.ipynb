```python
def plot(x, colors):
  
    palette = np.array(sb.color_palette("hls", 10))  #Choosing color palette 
    labels = ["z_1", "z_2", "z_3", "z_4"]

    # Create a scatter plot.
    f = plt.figure(figsize=(8, 8))
    ax = plt.subplot(aspect='equal')
    ax.set_xlim([x.min()*1.1, x.max()*1.1])
    ax.set_ylim([x.min()*1.1, x.max()*1.1])
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=40, c=palette[colors.astype(np.int)])
    # Add the labels for each digit.
    txts = []
    ax.set_xticks([])
    ax.set_yticks([])
    plt.axis('off')
    
    return f, ax, txts
```

```python
def plot_with_note(x, colors):
  
    palette = np.array(sb.color_palette("hls", 10))  #Choosing color palette 

    # Create a scatter plot.
    f = plt.figure(figsize=(16, 16))
    ax = plt.subplot(aspect='equal')
    ax.set_xlim([-30, 35])
    ax.set_ylim([-30, 30])
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=40, c=palette[colors.astype(np.int)])
    # Add the labels for each digit.
    txts = []
    for i in range(0, x.shape[0]):
        # Position of each label.
        xtext, ytext = x[i, 0], x[i, 1]
        txt = ax.text(xtext, ytext, str(i), fontsize=8)
        # txt.set_path_effects([pe.Stroke(linewidth=5, foreground="w"), pe.Normal()])
        txts.append(txt)
    
    return f, ax
```
